```python
def points_by_rank(rank: int):
    match rank:
        case 1:
            return 100
        case 2:
            return 80
        case 3:
            return 50
        case _:
            return 10
```

```python
def score_rule(group):
    return (
        group
        .sort_values(by='score', ascending=False)
        .drop_duplicates(subset=['user_address'], keep='first')
        .query('score > 0')
        .assign(rank=lambda x: range(1, x.shape[0] + 1))
        .assign(points=lambda x: x['rank'].apply(points_by_rank))
    )
```

```python
score_rule(c0)
```

```python
score_rule(c0)[['user_address', 'tape_id', 'proof', 'rank', 'points', 'score']]
```

```python
player_points = df.groupby('rule_id').apply(score_rule, include_groups=False).reset_index()
```

```python
pivoted = (
    pd
        .pivot(
        player_points,
        columns='rule_id',
        index='user_address',
        values=['points', 'rank', 'tape_id']
    )
    .assign(total_points=lambda x: x['points'].fillna(0.0).sum(axis=1))
    .sort_values(by='total_points', ascending=False)
    .swaplevel(axis=1).sort_index(axis=1, kind='stable', key=lambda x: x.map(lambda y: title_order.index(y) if y in title_order else -1))
)
```

```python
pivoted.index.name = None
pivoted.columns.names = [None, None]
```

```python
def format_user_address(addr: str):
    if addr in address_book:
        return address_book[addr]
    return addr[:8] + '...' + addr[-6:]


def format_title(title: str):
    if title in contests:
        return contests[title]
    return title.replace('_', ' ').title()


def format_tape(tape_id: str):
    return f'<a href="https://app.rives.io/tapes/{tape_id}">Tape</a>'

cols_points = [x for x in pivoted.columns if x[1] == 'points']
cols_rank = [x for x in pivoted.columns if x[1] == 'rank']
cols_tapes = [x for x in pivoted.columns if x[1] == 'tape_id']

(
    pivoted.style
    .format_index(format_title, axis=1)
    .format_index(format_user_address, axis=0)
    .format(na_rep='', precision=0)
    .format(format_tape, subset=cols_tapes, na_rep='')
    .set_table_styles({
        x: [
            {'selector': 'td', 'props': 'border-left: 1px solid white'}
        ]
        for x in cols_rank
    })
)
```
